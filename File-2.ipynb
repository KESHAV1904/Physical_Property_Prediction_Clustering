```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

acids = []


driver = webdriver.Chrome()  
driver.maximize_window()

for i in range(1, 30):
    url = 'https://www.sigmaaldrich.com/IN/en/products/chemistry-and-biochemicals/lab-chemicals/acids?country=IN&language=en&cmsRoute=products&cmsRoute=chemistry-and-biochemicals&cmsRoute=lab-chemicals&cmsRoute=acids&page='+str(i)
    print("Scraping page", i)
    
    driver.get(url)
    time.sleep(2)  
    
    page_source = driver.page_source
    
    soup = BeautifulSoup(page_source, 'html.parser')
    
    _class = soup.find_all('div', class_='jss209')
    
    for cls in _class:
        #print(cls)
        chemical_name = cls.find('span')
        #print(chemical_name)
        if chemical_name:
            acids.append(chemical_name.text)
            
driver.quit()

print(acids)
```

```python
salts = []

driver = webdriver.Chrome()  
driver.maximize_window()

for i in range(1, 77):
    url = f'https://www.sigmaaldrich.com/IN/en/products/chemistry-and-biochemicals/lab-chemicals/salts?country=IN&language=en&cmsRoute=products&cmsRoute=chemistry-and-biochemicals&cmsRoute=lab-chemicals&cmsRoute=salts&page={i}'
    print("Scraping page", i)
    
    driver.get(url)
    time.sleep(2) 
    
    page_source = driver.page_source
    
    soup = BeautifulSoup(page_source, 'html.parser')
    
    _class = soup.find_all('div', class_='jss198')
    
    for cls in _class:
        chemical_name = cls.find('span')
        if chemical_name:
            salts.append(chemical_name.text)

driver.quit()

print(salts)
```

```python
import pubchempy as pcp
no_chem = len(df)
formulae = []
for i in range(no_chem):
    print(i)
    string = df.iloc[i, 0]
    results = pcp.get_compounds(string, 'name')
    if results:
        formulae.append(results[0].isomeric_smiles)
    else:
        s = string.strip().split(' ')
        a = ' '.join(s[:len(s)-1])
        if a:
            results = pcp.get_compounds(a, 'name')
            if results:
                formulae.append(results[0].isomeric_smiles)
                df.iloc[i, 0] = a
            else:
                s = a.strip().split(' ')
                b = ' '.join(s[:len(s)-1])
                if b:
                    results = pcp.get_compounds(b, 'name')
                    if results:
                        formulae.append(results[0].isomeric_smiles)
                        df.iloc[i, 0] = b
                    else:
                        formulae.append(None)
                else:
                    formulae.append(None)
        else:
            formulae.append(None)
            
                
formulae
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

def get_molecular_properties(chemical_name, smiles):

    x = chemical_name.strip().split(' ')
    s= '+'.join(x)
    url = f'https://www.ncbi.nlm.nih.gov/pccompound/?term={s}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    url2_sec = soup.find('div', class_='rprt')
    if url2_sec:
        url2_fnd = url2_sec.find('a')
        url2 = url2_fnd.get('href')
        print(url2)
        driver = webdriver.Chrome()  
        driver.maximize_window()
        driver.get(url2)
        page_source = driver.page_source
        soup2 = BeautifulSoup(page_source, 'html.parser')
        #print(soup2)
        bp_sec = soup2.find('section', id = 'Boiling-Point')
        bp = None
        if bp_sec:
            bp_div = bp_sec.find('div', class_='break-words space-y-1')
            bp = bp_div.text.strip()
        mp_sec = soup2.find('section', id = 'Melting-Point')
        mp = None
        if mp_sec:
            mp_div = mp_sec.find('div', class_='break-words space-y-1')
            mp = mp_div.text.strip()
        sol_sec = soup2.find('section', id = 'Solubility')
        sol = None
        if sol_sec:
            sol_div = sol_sec.find('div', class_='break-words space-y-1')
            sol = sol_div.text.strip()
        driver.quit()
    else:
        mp = None
        bp = None
        sol = None
    print(mp, bp, sol)
    
    if smiles is None:
        s = [None] * 29
        return s + [mp, bp, sol]
    
    mol = Chem.MolFromSmiles(smiles)
    
    if mol is None:
        s = [None] * 29
        return s + [mp, bp, sol]
    
    properties = [
        Descriptors.MolWt(mol),
        Descriptors.ExactMolWt(mol),
        Descriptors.NumValenceElectrons(mol),
        Descriptors.NumRadicalElectrons(mol),
        Descriptors.MaxPartialCharge(mol),
        Descriptors.MinPartialCharge(mol),
        Descriptors.MaxAbsPartialCharge(mol),
        Descriptors.MinAbsPartialCharge(mol),
        Descriptors.FpDensityMorgan1(mol),
        Descriptors.FpDensityMorgan2(mol),
        Descriptors.FpDensityMorgan3(mol),
        Descriptors.BalabanJ(mol),
        Descriptors.BertzCT(mol),
        Descriptors.HallKierAlpha(mol),
        Descriptors.Kappa1(mol),
        Descriptors.Kappa2(mol),
        Descriptors.Kappa3(mol),
        Descriptors.Chi0(mol),
        Descriptors.Chi0n(mol),
        Descriptors.Chi0v(mol),
        Descriptors.Chi1(mol),
        Descriptors.Chi1n(mol),
        Descriptors.Chi1v(mol),
        Descriptors.Chi2n(mol),
        Descriptors.Chi2v(mol),
        Descriptors.Chi3n(mol),
        Descriptors.Chi3v(mol),
        Descriptors.Chi4n(mol),
        Descriptors.Chi4v(mol),
        mp,
        bp,
        sol
    ]
    print(properties)
    return properties
```
